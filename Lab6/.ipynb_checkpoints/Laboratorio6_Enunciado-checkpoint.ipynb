## 1.2 Creaci√≥n de nuevas Caracteristicas [2 Puntos] 

Como ya se les comento, Mr. Lepin esta interesado en obtener las caracter√≠sticas **LRMFP**, para esto les se√±ala que estas caracter√≠sticas se construyen en base a las siguientes definiciones:

- **Length (L)**: Intervalo de tiempo, en d√≠as, entre la primera y la √∫ltima visita del cliente. Mientras mas grande sea el valor, mas fiel es el cliente.
- **Recency (R)**: Indica la actualidad de la interacci√≥n de un cliente con la empresa, y da informaci√≥n sobre la tendencia a repetir la compra. Se define como: 
$$Recency(n)=\dfrac{1}{n} \sum^n_{i=1} date\_diff(t_{fecha final}, t_{m-i+1})$$

 Donde $date\_diff$ representa la diferencia en d√≠as entre la fecha de finalizaci√≥n del periodo de observaci√≥n ($t_{fecha final}$), y la fecha de una visita del cliente cercana a $t_{fecha final}$, $t_{m-i+1}; t_{m}$ es la √∫ltima visita del cliente; y n es el n√∫mero de visitas recientes del cliente consideradas.

- **Monetary (M)**: El t√©rmino "monetario" se refiere a la cantidad media de dinero gastada por cada visita del cliente durante el per√≠odo de observaci√≥n y refleja la contribuci√≥n del cliente a los ingresos de la empresa.

- **Frequency (F)**: Se refiere al n√∫mero total de visitas del cliente durante el periodo de observaci√≥n. Cuanto mayor sea la frecuencia, mayor ser√° la fidelidad del cliente. 

- **Periodicity (P)**: Representa si los clientes visitan las tiendas con regularidad.

$$Periodicity(n)=std(IVT_1, ..., IVT_n)$$

&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Donde $IVT$ denota el tiempo entre visitas y n representa el n√∫mero de valores de tiempo entre visitas de un cliente.
 

$$IVT_i=date\_diff(t_{i+1},t)$$

En base a las definiciones se√±aladas, dise√±e una funci√≥n que permita obtener las caracter√≠sticas **LRMFP** recibiendo un DataFrame como entrada. Para esto, no estar√° permitido el uso de iteradores, utilice todas las herramientas que les ofrece `pandas` para realizar esto.

Una referencia que le puede ser √∫til es el [documento original](https://www.researchgate.net/publication/315979555_LRFMP_model_for_customer_segmentation_in_the_grocery_retail_industry_a_case_study) en donde se propone este m√©todo.

**Nota:** Para la $fechafinal$ utilice la fecha m√°xima del dataset m√°s 1 d√≠a.

**Ejemplo de Resultado Esperado:**

```python
def period(x):
    '''funcion auxiliar para calcular periodicity''' 
    x = x["InvoiceDate"]
    dif = x - x.shift(1)
    dif.dropna(inplace=True)
    std_dif = dif.std()
    return std_dif

def custom_features(df_retail):
    
    # Para la  ùëìùëíùëê‚Ñéùëéùëìùëñùëõùëéùëô  utilice la fecha m√°xima del dataset m√°s 1 d√≠a.
    fecha_final = max(df_retail["InvoiceDate"]) + datetime.timedelta(days=1)
    
    # Length
    Length = pd.DataFrame((df_retail[["Customer ID","InvoiceDate"]].groupby("Customer ID").max() 
              - df_retail[["Customer ID","InvoiceDate"]].groupby("Customer ID").min())["InvoiceDate"].apply(lambda x: x.days))
    Length = Length.rename(columns={"InvoiceDate":"Length"})
    
    
    ### Pre procesamiento previo para siguientes features
    
    # agrega Total pagado
    df_retail_copy = df_retail.copy()
    df_retail_copy["Total"] = df_retail_copy["Quantity"]*df_retail_copy["Price"]
    
    # total pagado por invoice
    total_invoice = df_retail_copy.groupby(["Invoice"]).sum().reset_index(drop=False)[["Invoice","Total"]]
    
    # totales por invoice de cada cliente y fechas
    invoice_client = pd.merge(left=df_retail_copy[["Invoice","InvoiceDate","Customer ID"]],
                    right=total_invoice,
                    on="Invoice",
                    how="outer",
                    sort=True)
    invoice_client.drop_duplicates(subset = "Invoice", inplace=True, keep='first')
    
    # Recency
    invoice_client_copy = invoice_client.copy()
    invoice_client_copy["date_diff"] = (fecha_final - invoice_client_copy["InvoiceDate"]).apply(lambda x: x.days)
    Recency = pd.DataFrame(invoice_client_copy.groupby("Customer ID")["date_diff"].mean())
    Recency = Recency.rename(columns={"date_diff":"Recency"})
    
    # Monetary
    Monetary = invoice_client.groupby("Customer ID").mean()
    Monetary = Monetary.rename(columns={"Total":"Monetary"})
    
    # Frequency
    Frequency = pd.DataFrame(invoice_client.groupby("Customer ID").size())
    Frequency = Frequency.rename(columns={0:"Frecuency"})
    
    # Periodicity
    invoice_client_copy_2 = invoice_client.copy()
    Periodicity = pd.DataFrame(invoice_client_copy_2.groupby("Customer ID").apply(lambda x: period(x)))
    Periodicity = Periodicity.rename(columns={0:"Periodicity"})
    Periodicity["Periodicity"] = Periodicity["Periodicity"].apply(lambda x: x.days)
        
    ### Union de todas features para dataset final
    
    final_df = Length.merge(Recency,how="outer",left_index=True,right_index=True)
    final_df = final_df.merge(Monetary,how="outer",left_index=True,right_index=True)
    final_df = final_df.merge(Frequency,how="outer",left_index=True,right_index=True)
    final_df = final_df.merge(Periodicity,how="outer",left_index=True,right_index=True)
    
    # Para evitar dar una interpretaci√≥n errada, se eliminan los clientes con NaNs 
    final_df.dropna(inplace=True)
    
    return final_df
```

### 1.3.1 Estandarizar Caracteristicas [0.5 puntos]

Construya una clase llamada ``MinMax()`` utilizando ``BaseEstimator`` y ``TransformerMixin`` para realizar una transformaci√≥n de cada una de las columnas de un DataFrame utilizando ``ColumnTransformer()`` m√°s tarde (tome como referencia el siguiente [enlace](https://sklearn-template.readthedocs.io/en/latest/user_guide.html#transformer)).


 Para esto considere que Min-Max escaler queda dada por la ecuaci√≥n:

$$MinMax = \dfrac{x-min(x)}{max(x) - min(x)}$$

Con esto buscamos que los valores que componen a las columnas se muevan en el rango de valores $[0, 1]$.

**Respuesta:**

```python
class MinMax(BaseEstimator, TransformerMixin):
    
    def __init__(self):
        self
    
    def fit(self, X):
        self.max = []
        self.min = []
        for i in X.columns:
            self.max.append(max(X[i]))
            self.min.append(min(X[i]))
        return self
    
    def transform(self, X):
        X = X.copy()
        for index, feature in enumerate(X.columns):
            X[feature]=(X[feature]-self.min[index])/(self.max[index]-self.min[index])
        return X
```

### 1.3.2 `T-SNA` Pipeline [1.0 puntos]

Para comenzar introduci√©ndose en el uso de pipeline, decide probar realizando un pipeline enfocado en la reducci√≥n de dimensionalidad y as√≠ hacer no decepcionar a Mr. Lepin con la clusterizaci√≥n del modelo. 

Configure un pipeline utilizando el algoritmo `T-SNE` sobre los datos **LRMFP**, donde, para la realizaci√≥n del pipeline considera los siguientes pasos:

1. Como primer paso obtenga las caracter√≠sticas **LRMFP** desde el DataFrame ``df_retail_II_cleaned.pickle`` utilizando la funci√≥n ``custom_features`` creada anteriormente, junto a ``FunctionTransformer()``. Considere esto como el primer paso de su pipeline.
2. En segundo lugar usando ``ColumnTransformer()`` aplique el MinxMax scaler creado por usted sobre todas las columnas generadas en el paso anterior. 
3. Finalmente, aplique un √∫ltimo paso donde obtiene las 2 componentes m√°s relevantes utilizando el algoritmo T-sne de sckit-learn.

Tras aplicar las transformaciones sobre el dataset **LRMFP**, gr√°fique las componentes obtenidas en la reducci√≥n de dimensionalidad.

**Respuesta:**

```python
# Paso 1
LRMFP_features = FunctionTransformer(custom_features)
df_transform = LRMFP_features.transform(df_retail)

# Paso 2
preprocessing_transformer = ColumnTransformer(transformers=[('MinMax', MinMax(),  [i for i in df_transform.columns])])

# Paso 3
pipeline = Pipeline([('Preprocessing', preprocessing_transformer),
                     ('TSNE', TSNE(n_components=2, random_state=50))])
```

```python
# Ejecuci√≥n del pipeline
tsne_features = pipeline.fit_transform(df_transform)
tsne_features = pd.DataFrame(tsne_features,columns=['x_tsne','y_tsne'])
tsne_features
```

```python
# Gr√°fico de las componentes obtenidas en la reducci√≥n de dimensionalidad
fig = go.Figure()


fig.add_trace(go.Scatter(x=tsna_features['x_tsne'], y=tsna_features['y_tsne'],
                    mode='markers'))

fig.update_layout(
    title = "T-SNE",
    xaxis_title = 'x_tsne',
    yaxis_title = 'y_tsne',
    font=dict(
        size=12,
        color="black"
    )
)      
fig.show()
```
