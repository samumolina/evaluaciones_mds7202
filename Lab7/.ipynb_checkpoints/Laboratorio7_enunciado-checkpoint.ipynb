````python
### C√≥digo aqu√≠ ###
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, MinMaxScaler, OrdinalEncoder
````

````python
# Variables num√©ricas

# Variables num√©ricas que siguen una distribuci√≥n normal evidente (de acuerdo a reporte html)
normal_features = ['Height',
 'Weight',
 'Aggression',
 'Reactions',
 'Vision',
 'Composure',
 'Acceleration',
 'Speed',
 'Balance',
 'Jumping',
 ]

# Resto de num√©ricas
features_num = df_players.select_dtypes(exclude="object").columns.tolist()
for feature in normal_features:
    if feature in features_num:
        features_num.remove(feature)
    else:
        pass

# Variables categ√≥ricas sin considerar las relacionadas al Target y tambi√©n Name que ser√≠a irrelevante

features_cat = df_players.select_dtypes(include="object").columns.tolist()
features_cat.remove('Name')
features_cat.remove('National_Position')
features_cat.remove('Seleccionado')
````

    - Para variables num√©ricas que distribuyen normal (las m√°s evidentes) se utiliza StandardScaler.
    - Para el resto de num√©ricas se utiliza MinMaxScaler.
    - Para variables categ√≥ricas se utiliza OneHotEncoder

````python
# Columns transformer

ct = ColumnTransformer(transformers = 
                       [
                           
                        ('StandardScaler', StandardScaler(), normal_features), # para variables normales
                        ('MinMaxScaler', MinMaxScaler(), features_num), # resto de num√©ricas
                        ('OneHotEncoder', OneHotEncoder(handle_unknown="ignore"), features_cat), # categ√≥ricas
                           
                        ])
````

    De acuerdo a la gu√≠a, como tenemos menos de 100k de data, usaremos los clasificadores que siguen a continuaci√≥n:

````python
# Pipelines con diferentes clasificadores
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

# DecisionTree
tree_pipe = Pipeline(
    [("preprocesamiento", ct), ("tree", DecisionTreeClassifier(criterion="entropy"))]
)

# KNN
knn_pipe = Pipeline(
    [("preprocesamiento", ct), ("knn", KNeighborsClassifier(n_neighbors=3))]
)

# SVC lineal
svc_pipe = Pipeline(
    [("preprocesamiento", ct), ("svc", SVC(kernel='linear', probability=False))]
)
````

**Respuesta:**

````python
### C√≥digo aqu√≠ ###

#  Separaci√≥n del conjunto de datos en entrenamiento y de prueba

features = df_players.drop(columns=["Seleccionado"])
labels = df_players.loc[:, "Seleccionado"]

X_train, X_test, y_train, y_test = train_test_split(
    features, labels, test_size=0.33, shuffle=True, stratify=labels
)
````

````python
# Entrenamiento de los pipelines

tree_pipe.fit(X_train, y_train)
knn_pipe.fit(X_train, y_train)
svc_pipe.fit(X_train, y_train)
````

````python
# Predicciones

y_pred_tree = tree_pipe.predict(X_test)
y_pred_knn = knn_pipe.predict(X_test)
y_pred_svc = svc_pipe.predict(X_test)
````

````python
# M√©tricas para evaluar los modelos
print("Decision Tree")
print(classification_report(y_pred_tree, y_test))
print("")

print("KNN")
print(classification_report(y_pred_knn, y_test))
print("")

print("SVC lineal")
print(classification_report(y_pred_svc, y_test))
````

**Respuesta**:

```
A partir del reporte anterior, elos modelos muestran un alto accuracy (sobre 0.9), esto en principio no dice mucho ya que se debe principalmente a que las clases est√°n muy desequilibradas. Por lo anterior, esta no es una buena m√©trica para medir desempe√±os.
```

    De acuerdo a la gu√≠a, tenemos menos de 100k datos. A continuaci√≥n se ejecutan algunos modelos sugeridos:

````python
# Pipelines
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestRegressor

# Regresi√≥n lineal SOLO PARA COMPARAR, no es sugerencia de la gu√≠a
pipe = Pipeline([('preprocesamiento', preprocessing_transformer),
                 ('regresor', LinearRegression())])


# Modelos sugeridos de acuerdo a la gu√≠a
rigde_pipe = Pipeline([('preprocesamiento', preprocessing_transformer) , 
                       ('regresor', Ridge())])

lasso_pipe = Pipeline([('preprocesamiento', preprocessing_transformer) , 
                       ('regresor', Lasso(alpha=0.1))])

rf_pipe = Pipeline([("preprocesamiento", preprocessing_transformer) , 
                    ("Linear Reg", RandomForestRegressor(max_depth=15,
                                                         max_features=0.6,
                                                         n_estimators=500,
                                                         random_state=1))])
````

````python
# Separaci√≥n train y test

features = df.drop(columns=["Weekly Salary"])
labels = df.loc[:, "Weekly Salary"]

X_train, X_test, y_train, y_test = train_test_split(features, labels, 
                                                    test_size=0.33, shuffle=True)
````

````python
# Entrenamiento

pipe.fit(X_train, y_train)
rigde_pipe.fit(X_train, y_train)
lasso_pipe.fit(X_train, y_train)
rf_pipe.fit(X_train, y_train)
````

````python
# Predicciones

y_pred = pipe.predict(X_test)
ridge_y_pred = rigde_pipe.predict(X_test)
lasso_y_pred = lasso_pipe.predict(X_test)
rf_y_pred = rf_pipe.predict(X_test)
````

````python
# R¬≤ obtenidos
from sklearn.metrics import r2_score

print(f"R¬≤ Regresi√≥n Lineal: {r2_score(y_test, y_pred)}\n")
print(f"R¬≤ Regresi√≥n Ridge: {r2_score(y_test, ridge_y_pred)}\n")
print(f"R¬≤ Regresi√≥n Lasso: {r2_score(y_test, lasso_y_pred)}\n")
print(f"R¬≤ Regresi√≥n con RandomForestt: {r2_score(y_test, rf_y_pred)}")
````
