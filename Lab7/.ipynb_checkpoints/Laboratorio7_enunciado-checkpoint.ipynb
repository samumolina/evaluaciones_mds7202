````python
### Código aquí ###
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, MinMaxScaler, OrdinalEncoder
````

````python
# Variables numéricas

# Variables numéricas que siguen una distribución normal evidente (de acuerdo a reporte html)
normal_features = ['Height',
 'Weight',
 'Aggression',
 'Reactions',
 'Vision',
 'Composure',
 'Acceleration',
 'Speed',
 'Balance',
 'Jumping',
 ]

# Resto de numéricas
features_num = df_players.select_dtypes(exclude="object").columns.tolist()
for feature in normal_features:
    if feature in features_num:
        features_num.remove(feature)
    else:
        pass

# Variables categóricas sin considerar las relacionadas al Target y también Name que sería irrelevante

features_cat = df_players.select_dtypes(include="object").columns.tolist()
features_cat.remove('Name')
features_cat.remove('National_Position')
features_cat.remove('Seleccionado')
````

    - Para variables numéricas que distribuyen normal (las más evidentes) se utiliza StandardScaler.
    - Para el resto de numéricas se utiliza MinMaxScaler.
    - Para variables categóricas se utiliza OneHotEncoder

````python
# Columns transformer

ct = ColumnTransformer(transformers = 
                       [
                           
                        ('StandardScaler', StandardScaler(), normal_features), # para variables normales
                        ('MinMaxScaler', MinMaxScaler(), features_num), # resto de numéricas
                        ('OneHotEncoder', OneHotEncoder(handle_unknown="ignore"), features_cat), # categóricas
                           
                        ])
````

    De acuerdo a la guía, como tenemos menos de 100k de data, usaremos los clasificadores que siguen a continuación:

````python
# Pipelines con diferentes clasificadores
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

# DecisionTree
tree_pipe = Pipeline(
    [("preprocesamiento", ct), ("tree", DecisionTreeClassifier(criterion="entropy"))]
)

# KNN
knn_pipe = Pipeline(
    [("preprocesamiento", ct), ("knn", KNeighborsClassifier(n_neighbors=3))]
)

# SVC lineal
svc_pipe = Pipeline(
    [("preprocesamiento", ct), ("svc", SVC(kernel='linear', probability=False))]
)
````

Ahora, entrene los `pipeline` generados en los pasos anteriores. Para esto, primero separe los datos de entrenamiento en un conjunto de entrenamiento y de prueba  (la proporción queda a su juicio).

**Respuesta:**

````python
### Código aquí ###

#  Separación del conjunto de datos en entrenamiento y de prueba

features = df_players.drop(columns=["Seleccionado"])
labels = df_players.loc[:, "Seleccionado"]

X_train, X_test, y_train, y_test = train_test_split(
    features, labels, test_size=0.33, shuffle=True, stratify=labels
)
````

````python
# Entrenamiento de los pipelines

tree_pipe.fit(X_train, y_train)
knn_pipe.fit(X_train, y_train)
svc_pipe.fit(X_train, y_train)
````

````python
# Predicciones

y_pred_tree = tree_pipe.predict(X_test)
y_pred_knn = knn_pipe.predict(X_test)
y_pred_svc = svc_pipe.predict(X_test)
````

````python
# Se unen las bases de acuerdo a la columna Name que contiene los nombres de los jugadores

df = pd.merge(left=df_players,
         right=df_sueldos,
         on="Name",
         how="inner",
         sort=True)

df
````

    Solo 578 jugadores coincidian entre las bases de datos de jugadores y sueldos.

````python
# Se usarán solo VARIABLES NUMÉRCIAS

numericas = df.select_dtypes(exclude="object").columns.tolist()

# Se separan las que siguen distribución normal
for feature in normal_features:
    if feature in numericas :
        numericas.remove(feature)
    else:
        pass

# Se quita el target
numericas.remove("Weekly Salary")

# Columns transformer
preprocessing_transformer = ColumnTransformer(transformers = 
                       [
                           
                        ('StandardScaler', StandardScaler(), normal_features), # para variables normales
                        ('MinMaxScaler', MinMaxScaler(), numericas), # resto de numéricas
                         
                        ])
````

````python
# Pipeline
reg_pipe = Pipeline(
    [("preprocesamiento", preprocessing_transformer), ("Linear Reg", RandomForestRegressor(max_depth=15,
                                                                                           max_features=0.6,
                                                                                           n_estimators=500,random_state=1))]
)
````

    De acuerdo a la guía, tenemos menos de 100k datos. A continuación se ejecutan algunos modelos sugeridos:

````python
# Pipelines
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestRegressor

# Regresión lineal SOLO PARA COMPARAR, no es sugerencia de la guía
pipe = Pipeline([('preprocesamiento', preprocessing_transformer),
                 ('regresor', LinearRegression())])


# Modelos sugeridos de acuerdo a la guía
rigde_pipe = Pipeline([('preprocesamiento', preprocessing_transformer) , 
                       ('regresor', Ridge())])

lasso_pipe = Pipeline([('preprocesamiento', preprocessing_transformer) , 
                       ('regresor', Lasso(alpha=0.1))])

rf_pipe = Pipeline([("preprocesamiento", preprocessing_transformer) , 
                    ("Linear Reg", RandomForestRegressor(max_depth=15,
                                                         max_features=0.6,
                                                         n_estimators=500,
                                                         random_state=1))])
````

````python
# Separación train y test

features = df.drop(columns=["Weekly Salary"])
labels = df.loc[:, "Weekly Salary"]

X_train, X_test, y_train, y_test = train_test_split(features, labels, 
                                                    test_size=0.33, shuffle=True)
````

````python
# Entrenamiento

pipe.fit(X_train, y_train)
rigde_pipe.fit(X_train, y_train)
lasso_pipe.fit(X_train, y_train)
rf_pipe.fit(X_train, y_train)
````

````python
# Predicciones

y_pred = pipe.predict(X_test)
ridge_y_pred = rigde_pipe.predict(X_test)
lasso_y_pred = lasso_pipe.predict(X_test)
rf_y_pred = rf_pipe.predict(X_test)
````

````python
# R² obtenidos
from sklearn.metrics import r2_score

print(f"R² Regresión Lineal: {r2_score(y_test, y_pred)}\n")
print(f"R² Regresión Ridge: {r2_score(y_test, ridge_y_pred)}\n")
print(f"R² Regresión Lasso: {r2_score(y_test, lasso_y_pred)}\n")
print(f"R² Regresión con RandomForestt: {r2_score(y_test, rf_y_pred)}")
````

**Respuesta**:

```
El Coeficiente de Determinación R² es un puntaje que representa la proporción de varianza de los valores predichos por el modelo con respecto a los target reales. Permite entender la explicabilidad de la variable de interés a partir del las variables utilizadas para la predicción. Mientras más cercano a 1, mejor se explica el target con las variables independientes.

Para evitar interpretaciones posiblemente erradas respecto de las variables categóricas, se optó por ejecutar modelos solo usando las variables numéricas.
```
